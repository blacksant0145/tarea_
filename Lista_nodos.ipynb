```python

class NodoRopa:
    def __init__(self, nombre):
        self.nombre = nombre
        self.siguiente = None


class NodoInventario:
    def __init__(self, item, cantidad):
        self.item = item
        self.cantidad = cantidad
        self.siguiente = None


class NodoTikTok:
    def __init__(self, nombre_famoso, cantidad_tiktoksubidos):
        self.nombre_famoso = nombre_famoso
        self.cantidad_tiktoksubidos = cantidad_tiktoksubidos
        self.siguiente = None

# Función para crear y enlazar los nodos de ropa con ingreso de usuario
def crear_lista_ropa():
    camisa_nombre = input("Ingresa el nombre de la camisa: ")
    zapatos_nombre = input("Ingresa el nombre de los zapatos: ")
    pantalones_nombre = input("Ingresa el nombre de los pantalones: ")

    camisa = NodoRopa(camisa_nombre)
    zapatos = NodoRopa(zapatos_nombre)
    pantalones = NodoRopa(pantalones_nombre)

    camisa.siguiente = zapatos
    zapatos.siguiente = pantalones

    return camisa

# Función para crear y enlazar los nodos del inventario del aula con ingreso de usuario
def crear_lista_inventario():
    items = ["Escritorios", "Computadores", "Mouse", "Teclados", "Sillas", "Salón", "Tablero", "TV"]
    cabeza = None
    actual = None

    for item in items:
        cantidad = int(input(f"Ingresa la cantidad de {item}: "))
        nodo = NodoInventario(item, cantidad)

        if cabeza is None:
            cabeza = nodo
            actual = cabeza
        else:
            actual.siguiente = nodo
            actual = nodo

    return cabeza

# Función para crear y enlazar los nodos de TikTok con ingreso de usuario
def crear_lista_tiktok():
    cantidad_famosos = int(input("¿Cuántos famosos de TikTok quieres ingresar? "))
    cabeza = None
    actual = None

    for i in range(cantidad_famosos):
        nombre_famoso = input(f"Ingresa el nombre del famoso {i+1}: ")
        cantidad_tiktoksubidos = int(input(f"Ingrese la cantidad de TikToks subidos por {nombre_famoso}: "))
        nodo = NodoTikTok(nombre_famoso, cantidad_tiktoksubidos)

        if cabeza is None:
            cabeza = nodo
            actual = cabeza
        else:
            actual.siguiente = nodo
            actual = nodo

    return cabeza

# Función para imprimir la lista de ropa
def imprimir_lista_ropa(nodo):
    while nodo:
        print(f"Prenda: {nodo.nombre}")
        nodo = nodo.siguiente

# Función para imprimir la lista de inventario
def imprimir_lista_inventario(nodo):
    while nodo:
        print(f"Item: {nodo.item}, Cantidad: {nodo.cantidad}")
        nodo = nodo.siguiente

# Función para imprimir la lista de TikTok
def imprimir_lista_tiktok(nodo):
    while nodo:
        print(f"Famoso: {nodo.nombre_famoso}, Cantidad de TikToks: {nodo.cantidad_tiktoksubidos}")
        nodo = nodo.siguiente


print("Lista de Ropa:")
lista_ropa = crear_lista_ropa()
imprimir_lista_ropa(lista_ropa)

print("\nLista de Inventario del Aula:")
lista_inventario = crear_lista_inventario()
imprimir_lista_inventario(lista_inventario)

print("\nLista de Famosos en TikTok:")
lista_tiktok = crear_lista_tiktok()
imprimir_lista_tiktok(lista_tiktok)
```
